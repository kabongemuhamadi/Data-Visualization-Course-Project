```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn.datasets import load_boston #only using this import for getting the dataset,


boston = load_boston() #function used to retrieve the data set

data = boston.data

prices=boston.target[:50]

fig, ax = plt.subplots(nrows=3,ncols=1)

crime_data = data[:,0][:50]
avg_rooms = data[:,5][:50]
nitrous_oxide = data[:,4][:50]
pupil_teacher_ratio = data[:,-4][:50]

x = [x for x in range(50)]

ax[0].plot()
ax[1].plot()
ax[2].plot()

ax[1].hist(avg_rooms)
ax[2].scatter(nitrous_oxide,pupil_teacher_ratio)
ax[0].plot(x, crime_data, prices)

fig.suptitle("Analysis of Boston Housing Market Data in 1970's")
ax[0].set_title("Crime Data vs Housing Prices Over Records") 
ax[1].set_title("Count of Average Rooms per Home by Neighborhood") 
ax[2].set_title("Nitrous Oxide vs Pupil Teacher Ratio") 
```
